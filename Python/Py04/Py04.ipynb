```python
#프로젝트 1번

#: 직사각형, 평행사변형, 사다리꼴 넓이를 구하는 메서드를 직접 작성

# 넓이를 구하려면 데이터(가로 x 세로)가 필요
# 함수보단 클래스(데이터와 기능을 하나로)
#직사각형, 평행사변형, 사다리꼴 모두 사각형(클래스와 상속 구조)

#1.클래스 구현
#2.생성자 만들기
#3.인스턴스 생성
#4.문자열 메서드
#5. 상속
```

```python
# 1.사각형 클래스 만들기

class Square:
    def __init__(self):
        # 사용자가 입력한 가로, 세로 값을 담을 리스트 (인스턴스 속성)
        self.square = []

    def input_side(self):
        # 가로, 세로 길이를 입력받아 리스트에 저장하는 공통 메서드
        width = int(input("가로 길이를 입력하세요: "))
        height = int(input("세로 길이를 입력하세요: "))
        self.square = [width, height]
# 2. 자식 클래스: 직사각형 (Rectangle)
class Rectangle(Square):
    def area(self):
        # 가로 * 세로
        result = self.square[0] * self.square[1]
        return f"직사각형의 넓이는 {result}입니다."
# 3. 자식 클래스: 평행사변형 (Parallelogram)
class Parallelogram(Square):
    def area(self):
        # 밑변 * 높이 (입력받은 가로, 세로를 각각 밑변, 높이로 간주)
        result = self.square[0] * self.square[1]
        return f"평행사변형의 넓이는 {result}입니다."
# 4. 자식 클래스: 사다리꼴 (Trapezoid)
class Trapezoid(Square):
    def input_side(self):
        # [오버라이딩] 사다리꼴은 밑변, 윗변, 높이 3개가 필요하므로 새로 정의합니다.
        base = int(input("밑변 길이를 입력하세요: "))
        top = int(input("윗변 길이를 입력하세요: "))
        height = int(input("높이 길이를 입력하세요: "))
        self.square = [base, top, height]
        
    def area(self):
        # (밑변 + 윗변) * 높이 / 2
        result = (self.square[0] + self.square[1]) * self.square[2] / 2
        return f"사다리꼴의 넓이는 {result}입니다."
```

```python
# 1. 어떤 도형을 계산할지 선택받기
print("=== 도형 넓이 계산기 ===")
menu = int(input("넓이를 구하고 싶은 사각형의 숫자를 입력하세요.\n1. 직사각형  2. 평행사변형  3. 사다리꼴\n>>> "))

# 2. 선택한 번호에 맞는 객체 생성
if menu == 1:
    shape = Rectangle()
    print("\n[직사각형 모드]")
elif menu == 2:
    shape = Parallelogram()
    print("\n[평행사변형 모드]")
elif menu == 3:
    shape = Trapezoid()
    print("\n[사다리꼴 모드]")
else:
    print("잘못된 번호입니다. 1, 2, 3 중에서 골라주세요.")
    shape = None

# 3. 객체가 정상적으로 생성되었다면 메서드 실행
if shape:
    shape.input_side()  # 길이를 입력받는 창이 뜹니다.
    print("-" * 30)
    print(shape.area())   # 계산된 넓이를 출력합니다.
    print("-" * 30)
```

#프로젝트 2

#1. Kiosk 클래스를 생성하고, 주문, 추가 주문 메서드를 적절하게 구현
#- Kiosk 클래스를 먼저 생성
#- 메뉴 출력 메서드
#- 주문 메서드
#- 추가주문 메서드

#2. 지불, 주문표 작성 메서드를 적절하게 구현


```python
class Kiosk:    
    def __init__(self):
        # 문제 2-1: 메뉴와 가격 리스트 초기화
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []    # 주문한 메뉴 이름을 담을 빈 리스트
        self.order_price = []   # 주문한 메뉴 가격을 담을 빈 리스트
    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')
   # 주문 메서드
    def menu_select(self):
        print()  # 줄 바꿈
        n = 0
        # 1. 메뉴 번호 선택
        while n < 1 or len(self.menu) < n:
            n = int(input("음료 번호를 입력하세요 : ")) 

            if 1 <= n and n <= len(self.menu):
                self.price_sum = self.price[n-1] 
                self.selected_menu = self.menu[n-1]
                print(self.selected_menu, ' : ', self.price_sum, '원') 
            else:  
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 2. 온도 선택
        t = 0
        while t != 1 and t != 2:
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:    
                print("1과 2 중 하나를 입력하세요.\n")

        # 3. [핵심] 선택한 메뉴를 장바구니(리스트)에 추가하기
        # 이 코드가 있어야 pay() 메서드에서 합계가 0원이 나오지 않습니다.
        self.order_menu.append(self.temp + ' ' + self.selected_menu)
        self.order_price.append(self.price_sum)
        
        print(f'>>> {self.temp} {self.selected_menu}가 장바구니에 담겼습니다.')
    # 추가 주문 메서드
    def another_order(self):
        while True:
            print("-" * 30)
            # 사용자에게 추가 주문 여부를 묻습니다.
            choice = input("추가 주문하시겠습니까? (y/n) : ").lower()
            
            # 'y' 또는 'yes'를 입력하면 다시 메뉴 선택으로 이동
            if choice == 'y' or choice == 'yes':
                self.menu_select()
                
            # 'n' 또는 'no'를 입력하면 추가 주문 루프 종료
            elif choice == 'n' or choice == 'no':
                print("추가 주문을 종료합니다.")
                break
                
            # 이상한 글자를 입력했을 때
            else:
                print("y 또는 n으로 입력해주세요.")
    def pay(self):
        print("\n" + "*" * 20)
        print("      주문 내역서      ")
        print("*" * 20)
        
        # 1. 장바구니(self.order_menu)에 담긴 메뉴와 가격을 하나씩 출력합니다.
        for i in range(len(self.order_menu)):
            print(f"{i+1}. {self.order_menu[i]} : {self.order_price[i]}원")
        
        # 2. 주문한 모든 메뉴의 합계를 구합니다.
        total_price = sum(self.order_price)
        print("-" * 25)
        print(f"합계 금액 : {total_price}원")
        print("-" * 25)
        
        # 3. 결제 금액 입력받기 및 잔돈 계산
        payment = 0
        while payment < total_price:
            money = int(input(f"결제하실 금액을 입력하세요 (총 {total_price}원) : "))
            payment += money
            
            if payment < total_price:
                print(f"금액이 {total_price - payment}원 부족합니다. 더 입력해주세요.")
        
        # 4. 결제 완료 및 잔돈 출력
        print(f"\n결제가 완료되었습니다!")
        print(f"거스름돈은 {payment - total_price}원입니다. 이용해주셔서 감사합니다.")
```

```python
# 1. 'cafe'라는 이름의 키오스크 기계 설치 (인스턴스 생성)
cafe = Kiosk()

# 2. 손님에게 메뉴판 보여주기
print("="*10, " WELCOME TO CAFE ", "="*10)
cafe.menu_print()

# 3. 첫 번째 주문 받기 (메뉴 고르고 -> 온도 고르고)
cafe.menu_select()

# 4. "더 주문하시겠어요?" 물어보기 (y 누르면 3번 반복, n 누르면 종료)
cafe.another_order()

# 5. [중요!] 주문이 모두 끝났으니 이제 영수증 보여주고 결제하기
cafe.pay()
```
