```python
# 은행계좌 만들기
#1. Account 클래스 = 전체 시스템의 핵심 설계
#2. 속성 = 은행명 / 예금주 / 계좌번호 / 잔액
#3. 생성자 만들기 : 입력 + 자동 생성 섞인 구조
#4. self로 인스턴스 속성 만들기
#5. 예금주 + 초기잔액만 입력
```

```python
import random

class Account:  # Q1: Account 클래스 정의
    count = 0   #  Q2: 생성된 계좌 수를 저장하는 클래스 변수

    def __init__(self, name, balance):
        self.bank = "SC은행"
        self.name = name          # 예금주
        self.balance = balance   # 잔액
        self.deposit_count = 0   #  Q6: 입금 횟수 초기화
        self.account_num = self.make_account_number()  #  Q3: 계좌번호 생성

        Account.count += 1  # Q2: 계좌 생성 시 count 증가

    #  Q3: 계좌번호 랜덤 생성 메서드
    def make_account_number(self):
        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)
        return f"{num1}-{num2}-{num3}"

    #  Q4: 입금 메서드
    def deposit(self, amount):
        if amount < 1:  # 최소 1원 조건
            print("입금은 최소 1원 이상만 가능합니다.")
            return

        self.balance += amount
        self.deposit_count += 1  #  Q6: 입금 횟수 증가

        print(f"[입금] {amount:,}원 → 잔액: {self.balance:,}원")

        self.check_interest()  #  Q7: 이자 로직 호출

    #  Q7: 이자 지급 메서드 (5회마다 1%)
    def check_interest(self):
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            print(f"🎁 이자 지급! (+{interest:,}원) → 잔액: {self.balance:,}원")

    # Q5: 출금 메서드
    def withdraw(self, amount):
        if amount < 1:
            print("출금은 최소 1원 이상만 가능합니다.")
            return

        if amount > self.balance:
            print("잔액이 부족합니다.")
            return

        self.balance -= amount
        print(f"[출금] {amount:,}원 → 잔액: {self.balance:,}원")

    # Q8: 잔액 출력 (3자리 콤마)
    def display_balance(self):
        print(f"현재 잔액: {self.balance:,}원")

    #  Q9: 전체 계좌 개수 반환 (클래스 메서드)
    @classmethod
    def get_account_num(cls):
        return cls.count
```

```python
accounts = [] #Q10

acc1 = Account("A", 100000000)
acc2 = Account("B", 5000)
acc3 = Account("C", 200000000)

accounts.append(acc1)
accounts.append(acc2)
accounts.append(acc3)

# 1억 이상 계좌 출력
for acc in accounts:
    if acc.balance >= 100000000:
        print(f"[고액계좌] {acc.name} / {acc.balance:,}원")
```
